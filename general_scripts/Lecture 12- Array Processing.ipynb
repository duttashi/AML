```python
# Usage of numpy where

A= np.array([1,2,3,4])

B= np.array([10,20,30,40])

```

```python
condition=np.array([True,True,False,False])
```

```python
answer= [(A_val if cond else B_val) for A_val,B_val,cond in zip(A,B,condition)] # will display show me A_val if cond is True else show B_val
```

```python
answer
```

```python
# Note: The code in line 15 can be replaced by using the where clause in numpy as given
answer2= np.where(condition,A,B)
```

```python
answer2
```
